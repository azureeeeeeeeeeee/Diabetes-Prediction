```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Glucose', y='BloodPressure')
plt.title('Scatter Plot for Glucose and BloodPressure')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Pregnancies', y='Insulin')
plt.title('Scatter Plot for Pregnancies and Insulin')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Age', y='Outcome')
plt.title('Scatter Plot for Age and Outcome')
plt.show()
```

We can see some outliers, lets remove :
- All `Glucose` that's' <50
- All `BloodPressire` that's <40
- All `Insulin` that's' <0 and >400
- All `Pregnancies` that's >10
- All `Age` that's >70

# Data Cleaning

```python
df0 = df.copy()
df0.describe()
```

```python
df1 = df0[df0.Glucose > 50]
```

```python
df2 = df1[df1.BloodPressure > 40]
```

```python
df3 = df2[df2.Insulin != 0]
df3 = df3[df3.Insulin < 500]
df3.describe()
```

```python
df4 = df3[df3.Pregnancies < 10]
df4 = df4[df4.Age <= 70]
```

## Lets Do Visualization Again

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df4, x='Glucose', y='BloodPressure')
plt.title('Scatter Plot for Glucose and BloodPressure')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df4, x='Pregnancies', y='Insulin')
plt.title('Scatter Plot for Pregnancies and Insulin')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df4, x='Age', y='Outcome')
plt.title('Scatter Plot for Blood Pressure and Outcome')
plt.show()
```
